```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Sample - Superstore.csv', encoding='latin-1')
df
```

```python
df.isna().count()
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
inplace=True
```

```python
df.info()
```

```python
df.head()
```

```python
df['Order month'] = df['Order Date'].dt.month
df['Order date'] = df['Order Date'].dt.year
df['order day of week'] = df['Order Date'].dt.dayofweek
```

```python
df.head(10)
```

```python
sales_by_month = df.groupby('Order month')['Sales'].sum()
```

**Sales By Month**

```python
sales_by_month.plot(x='month', y='Sales', title='Sales By Month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

**Sales By Category**

```python
sales_by_category = df.groupby('Category')['Sales'].sum()
```

```python

plt.pie(sales_by_category, labels=sales_by_category.index, autopct='%1.1f%%')
plt.show()
```

**Sales By Sub-Category**

```python
df.head()
```

```python
sales_by_subCategory = df.groupby('')['Sales'].sum()
sales_by_subCategory
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x=sales_by_subCategory.index, y=sales_by_subCategory.values)
plt.xlabel('sub-Category')
plt.ylabel('sales')
plt.show()
```
